```python
# Open Route Service API
ors_key = "5b3ce3597851110001cf6248e448c057fc6a4b69a2ba0a7a30f2a2fa"

client = ors.Client(key=ors_key)
```

```python
# Coordinates is [lon, lat]
coordinates = [[110.392422455923, -7.83346560376681], [110.389955, -7.82794], [110.361905, -7.801278]]

# Direction
route = client.directions(coordinates=coordinates,
                          profile='driving-car',
                          format='geojson')
```

```python
route['features'][0]['properties']['summary']['distance']
```

```python
coords = route['features'][0]['geometry']['coordinates']

coords
```

```python
new_coords = [[coord[1], coord[0]] for coord in coords]

new_coords
```

```python
# Map
map_directions = folium.Map(location=[-7.797068, 110.370529], zoom_start=12)

# Creating feature groups
f1 = folium.FeatureGroup("Vehicle 1")

# Adding lines to the different feature groups
line_1 = folium.vector_layers.PolyLine(new_coords, popup='<b>Path of Vehicle_1</b>',tooltip='Vehicle_1',color='red',weight=4).add_to(f1)

f1.add_to(map_directions)

# Create marker
folium.Marker(location=new_coords[0],popup='Custom Marker 1',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='red',icon='none')).add_to(map_directions)
folium.Marker(location=new_coords[-1],popup='Custom Marker 2',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='red',icon='none')).add_to(map_directions)

folium.LayerControl().add_to(map_directions)

map_directions
```
